Data Represent and Tidying

```python
def tidy_up_line(line):
    '''
    tidy up the line with answer from ubotu. 
    Removes unecessary things like timestamp and <username>
    Example:
        2007-02-13-ubuntu.txt:[08:37] 
        <ubotu> benjm: Don't ask to ask a question. Just ask your question :)
        reformats this into "Don't ask to ask a question. Just ask your question :)"
    '''
#     removes timestamp
    first_space_idx = line.find(" ")
    line = line[first_space_idx+1: ]
#     removea username
    right_bracket = line.find(">")
    line = line[right_bracket+2: ]
    return line.lower()
```

```python
def build_label_to_ans_dict(filepath):
    '''
    hash !label to answer and save it to answer
    '''
#     python dict
    label_to_ans = {}
    with open(filepath, 'r') as f:  
        for line in f:
            line = line.strip()
            nextLine = f.readline()
            if not nextLine:
                break
            if nextLine.find('<ubotu>') != -1:
                nextLine = nextLine.strip()
                index_of_exclamation = line.find("!")
                index_of_space = line.find(" ", index_of_exclamation)
                label = line[index_of_exclamation: index_of_space]
                label_to_ans[label] = tidy_up_line(nextLine)
    with open('label_to_bot_answers.txt', 'w') as outfile:
        json.dump(label_to_ans, outfile, sort_keys=True, indent=4)
```

```python
def build_label_frequecy(label_list):
    label_to_freq = {}
    for label in label_list:
        if label in label_to_freq:
            label_to_freq[label] += 1
        else:
            label_to_freq[label] = 1
    return label_to_freq
```

```python
def build_sentence_label_pair_list(filepath):
    '''
    Example:
    2007-02-14-ubuntu.txt:[05:10] <kenthomson> I need a widescreen 1440x900 (16:10), GFXBOOT THEME. 
    I can't find it anywhere on the internet, can someone point me to a source?
    2007-02-14-ubuntu.txt:[05:10] <rambo3> !ot

    ("I need a widescreen 1440x900 (16:10), GFXBOOT THEME.
    I can't find it anywhere on the internet, can someone point me to a source?", "!ot")
    '''
    sentence_label_pair_list = []
    with open(filepath, 'r') as f:  
        for line in f:
            potential_key = f.readline()
            if not potential_key:
                break
            potential_answer_by_ubotu = f.readline()
            if not potential_answer_by_ubotu:
                break
            if potential_answer_by_ubotu.find('<ubotu>') != -1:
                line = line.strip()
                line = tidy_up_line(line)
                index_of_exclamation = potential_key.find("!")
                index_of_space = potential_key.find(" ", index_of_exclamation)
                label = potential_key[index_of_exclamation: index_of_space] 
                sentence_label_pair_list.append((line, label))
    return sentence_label_pair_list
```

```python
def unique_word_dict(sentence_label_pair_list):
    idx = 0
    unique_dict = {}
    for pair in sentence_label_pair_list:
        sentence = pair[0]
        for punctuation in string.punctuation:
            sentence = sentence.replace(punctuation, " ")
        sentence = sentence.lower()
        splitted_list = sentence.split()
        for word in splitted_list:
            if word not in unique_dict:
                unique_dict[word] = idx
                idx+=1
    return unique_dict
```

```python
def one_hot_encoding(X, dictionary):
    num_of_features = len(X)
    num_of_words = len(dictionary)
    feature_matrix = np.zeros((num_of_features, num_of_words))
    feature_idx = 0
    for sentence in X:
        sentence = sentence.lower()
        split_sen = sentence.split()
        for word in split_sen:
            if word in dictionary:
                feature_matrix[feature_idx, dictionary[word]] = 1
        feature_idx += 1
    return feature_matrix
```

```python
def select_labels(label_to_freq):
    sorted_dict_items = sorted(label_to_freq.items(), key=lambda d: d[1], reverse=True)
    selected_labels = []
    for i in range(0, 16):
        selected_labels.append(sorted_dict_items[i][0])
    return selected_labels
#  keep the ones with 15 labels, experiment with different amount
```

```python
filepath = "data-full.txt"
# build label to answer of the full dataset
# create label_pair_list, label_list out of the whole dataset
sentence_label_pair_list_whole, label_list =  build_sentence_label_pair_list(filepath)
sentence_label_pair_list_selected = []
label_to_freq_whole = build_label_frequecy(label_list)
selected_label_whole = select_labels(label_to_freq_whole)
f = open("selected.txt", "w")
for pair in sentence_label_pair_list_whole:
    if pair[1] in selected_label_whole:
        sentence_label_pair_list_selected.append((pair[0], pair[1]))
        f.write(pair[0]+"\n")
f.close()
print(sentence_label_pair_list_selected)
```

```python
uniq_word_dict =  unique_word_dict(sentence_label_pair_list_selected)
train_len = math.floor(0.8 * len(sentence_label_pair_list_selected))
val_len = math.floor(0.1 * len(sentence_label_pair_list_selected))
test_len = math.floor(0.1 * len(sentence_label_pair_list_selected))

train_data = []
val_data = []
test_data = []
f = open("train.txt", "w")
for idx in range(0, train_len):
    if sentence_label_pair_list_selected[idx][1] != '':
        str_tmp = "__label__"  + sentence_label_pair_list_selected[idx][1]+ " "+ sentence_label_pair_list_selected[idx][0] + "\n"
        f.write(str_tmp)
        train_data.append(sentence_label_pair_list_selected[idx])
f.close()

f = open("validation.txt", "w")
for idx in range(train_len+1, train_len + 1 + val_len):
    if sentence_label_pair_list_selected[idx][1] != '':
        str_tmp = "__label__" + sentence_label_pair_list_selected[idx][1] + " "+ sentence_label_pair_list_selected[idx][0]+ "\n"
        f.write(str_tmp)
        val_data.append(sentence_label_pair_list_selected[idx])
f.close()
    
f = open("test.txt", "w")
for idx in range(train_len + 1 + val_len, train_len+ val_len + 1 + test_len):
    if sentence_label_pair_list_selected[idx][1] != '':
        str_tmp = "__label__" + sentence_label_pair_list_selected[idx][1] + " "+ sentence_label_pair_list_selected[idx][0]+ "\n"
        test_data.append(sentence_label_pair_list_selected[idx])
        f.write(str_tmp)
f.close()
    
X_test = []
Y_test = []

for pair in test_data:
    X_test.append(pair[0])
    Y_test.append(pair[1])
    
X_train = []
Y_train = []

for pair in train_data:
    X_train.append(pair[0])
    Y_train.append(pair[1])
    
X_val = []
Y_val = []

for pair in val_data:
    X_val.append(pair[0])
    Y_val.append(pair[1])
```

```python
def zero_one_loss(y_pred, y_true):
    score = 0
    assert(len(y_pred) == len(y_true))
    for idx in range(len(y_pred)):
        if y_pred[idx] == y_true[idx]:
            score += 1
    return score/len(y_true)
```
